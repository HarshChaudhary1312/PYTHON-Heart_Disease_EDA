# Distribution Plot for Numeric Columns

```python
# Distributon of age & plotting multiple graph types at same time
sb.distplot(df['Age'], kde=True, hist=True, rug=True, color='Blue')
```

```python
# Divide the dataset into 15 bins
sb.distplot(df['Age'], bins=15)
```

# Correlation of Heatmap
 Correlation ranges btw 1 to -1 more closer to 1 better correlation between 2 columns<br>
 Closer to -1 denotes negative correlation that means if one increases other decreases<br>
 It defines the dependency btw 2 columns, like if one values fluctuates then how other value is affected


```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr()
```

```python
# Let's select the columns with datatype as numbers
num_df=df.select_dtypes(include=['number'])
num_df
```

```python
num_df.corr()
```

```python
sb.heatmap(num_df.corr())
```

# Correlation - JointPlot

```python
sb.jointplot(y='Age', x='MaxHR', data=df)
```

```python
# 100% Correlation
sb.jointplot(y='Age', x='Age', data=df)
```

```python
sb.jointplot(y='Oldpeak', x='Cholesterol', data=df)
```

```python
# Line doesn't fit the data : no correlation
sb.jointplot(x='Cholesterol',y='FastingBS',kind='reg',data=df)
```

```python
sb.jointplot(y='Age', x='MaxHR', data=df, kind='hex')
```

# Correlation- PairPlot
Jointplots are useful for visualizing the relationship between two variables and the individual distributions of the variables. They can also be used to identify patterns in the data and to estimate the strength of the relationship between the variables.
